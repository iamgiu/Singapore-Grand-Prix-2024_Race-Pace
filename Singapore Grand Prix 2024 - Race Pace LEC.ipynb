# Compare the race pace of Leclerc with Norris, Verstappen and Piastri

```python
import fastf1
from fastf1 import plotting
from fastf1 import utils

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
race = fastf1.get_session(2024, 'Singapore', 'R')
race.load()
```

```python
race.event
```

```python
lec = race.laps.pick_driver('LEC')
nor = race.laps.pick_driver('NOR')
ver = race.laps.pick_driver('VER')
pia = race.laps.pick_driver('PIA')
```

```python
fastf1.plotting.setup_mpl(mpl_timedelta_support=True)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(ver['LapNumber'], lec['LapTime'], color='red')
ax.plot(nor['LapNumber'], nor['LapTime'], color='orange')
ax.plot(ver['LapNumber'], ver['LapTime'], color='blue')
ax.plot(ver['LapNumber'], pia['LapTime'], color='yellow')
ax.set_title("LEC vs NOR vs VER vs PIA")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
plt.show()
```

Red Line (LEC):
LEC has a high lap time in the early laps but quickly stabilizes around 1:39.
After the pit stop (arount lap 32), LECs lap times return to a similar range as before, though with some variability towards the end.

Blue Line (VER):
VER starts with consistently low lap times and gradually lowers his lap times to around 1:361:37.
Following the spike (the pit stop), VER maintains a consistent pace, mostly with lap times between 1:36 and 1:37, indicating strong, steady performance.

Orange Line (NOR):
NOR has a similar trend as VER, with a steady start around 1:39 before lowering his lap times slightly.
NORs lap times after the spike remain stable but slightly higher compared to VER, showing he is consistently slower by a small margin.

Yellow Line (PIA):
PIA shows similar early lap times as NOR and VER, with lap times hovering around 1:39 initially.
Post-spike, PIAs lap times stabilize close to those of NOR, though with some variability near the end of the race.

Key Observations:
Pit Stops/Incidents: All drivers have visible spikes around laps 3035. These large increases in lap times likely indicate pit stops or other incidents affecting their lap times.
Consistency: VER (blue line) appears to be the most consistent after the initial laps and has the lowest average lap time, suggesting he was the fastest of the four drivers over the race.

Race Phases:
Start: All drivers have a slightly slower first lap as they settle into the race.
Middle: Between laps 5 and 30, VER (blue) and NOR (orange) show slight improvement, maintaining a consistent pace around 1:361:39, while LEC and PIA have occasional fluctuations.
End: After the spikes around lap 30-35, all drivers settle into a consistent pace, but VER appears to maintain the lead with consistently lower lap times than the others.

Summary:
VER (blue) likely had the strongest race pace and showed the most consistency across the laps.
NOR (orange) and PIA (yellow) followed a similar pattern but were generally slower than VER.
LEC (red) had more variability in lap times, suggesting potential challenges with consistency or additional pit stops or incidents.
This graph shows how each drivers pace evolved throughout the race, with VER appearing to dominate based on consistent, faster lap times.

```python
data={'Compound':lec['Compound'],'LapTime':lec['LapTime'],'Stint':lec['Stint']}
leclerc=pd.DataFrame(data)
```

```python
data={'Compound':nor['Compound'],'LapTime':nor['LapTime'],'Stint':nor['Stint']}
norris=pd.DataFrame(data)
```

```python
data={'Compound':ver['Compound'],'LapTime':ver['LapTime'],'Stint':ver['Stint']}
verstappen=pd.DataFrame(data)
```

```python
data={'Compound':pia['Compound'],'LapTime':pia['LapTime'],'Stint':pia['Stint']}
piastri=pd.DataFrame(data)
```

```python
leclerc=leclerc[leclerc['LapTime'] <= '0 days 00:01:40.00000']
norris=norris[norris['LapTime'] <= '0 days 00:01:40.00000']
verstappen=verstappen[verstappen['LapTime'] <= '0 days 00:01:40.00000']
piastri=piastri[piastri['LapTime'] <= '0 days 00:01:40.00000']
```

```python
leclerc_1stMedium_stint=leclerc['LapTime'].loc[leclerc['Stint']==1]
leclerc_2ndHard_stint=leclerc['LapTime'].loc[leclerc['Stint']==2]

norris_1stMedium_stint=norris['LapTime'].loc[norris['Stint']==1]
norris_2ndHard_stint=norris['LapTime'].loc[norris['Stint']==2]

verstappen_1stMedium_stint=verstappen['LapTime'].loc[verstappen['Stint']==1]
verstappen_2ndHard_stint=verstappen['LapTime'].loc[verstappen['Stint']==2]

piastri_1stMedium_stint=piastri['LapTime'].loc[piastri['Stint']==1]
piastri_2ndHard_stint=piastri['LapTime'].loc[piastri['Stint']==2]
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

columns=[leclerc['LapTime'],norris['LapTime'],verstappen['LapTime'],piastri['LapTime']]
ax.set_title("Race Pace")
ax.boxplot(columns,boxprops={'color':'white'},flierprops={'markeredgecolor':'white'},medianprops={'color':'white'},whiskerprops={'color':'white'},capprops={'color':'white'})
plt.xticks([1,2,3,4], ["LEC Avg Pace","NOR Avg Pace","VER Avg Pace","PIA Avg PACE"])
plt.show()
```

Leclerc demonstrated competitive pace potential, especially in his faster laps, which were on par with or even faster than the other drivers except Norris. However, his average race pace (as shown by the median) was not better than Norriss due to variability likely caused by traffic and other factors.

If Leclerc had not been impeded by slower cars, he might have maintained a more consistent pace closer to Norris. Overall, Norris had the best pace in terms of both speed and consistency, while Leclerc had flashes of comparable speed but was affected by race conditions.

1. Leclerc (LEC)
- 1st Stint (Medium Tires):
    The median lap time for Leclerc on mediums is just below 1:39.0, making it one of the slower median times compared to others in the first stint.
    He shows a fairly large interquartile range (IQR), indicating a wide variability in lap times, which suggests that he struggled to maintain a consistent pace on the mediums.
- 2nd Stint (Hard Tires):
    Leclercs median pace on hards is closer to 1:37.5, showing a significant improvement compared to his first stint.
    However, the IQR in this stint is still wide, meaning he experienced variability in lap times despite the improved pace.
    This suggests that Leclerc was generally faster on hards but still faced challenges in achieving consistent lap times.

2. Norris (NOR)
- 1st Stint (Medium Tires):
    Norris has a fast median lap time, around 1:37.0, with a narrow IQR, indicating strong consistency and speed on the medium tires.
    He outperformed the other drivers in this stint, showing that he was comfortable and effective on the mediums.
- 2nd Stint (Hard Tires):
    On the hards, Norris continues to show a narrow IQR, and his median lap time remains competitive, just above 1:37.0.
    His consistency and pace remain strong, indicating that Norris adapted well to both tire compounds and was able to sustain his quick and consistent performance throughout the race.

3. Verstappen (VER)
- 1st Stint (Medium Tires):
    Verstappens median time on mediums is slightly below 1:38.0, indicating he had a good pace, though not as fast as Norris.
    His IQR is narrower than Leclercs but wider than Norriss, suggesting moderate consistency with a solid but not exceptional pace.
- 2nd Stint (Hard Tires):
    On the hards, Verstappens median lap time improves slightly, close to 1:37.0, and his IQR is very narrow, showing strong consistency.
    This implies Verstappen was able to maintain a steady pace on hards, comparable to Norris, though Norris still showed a slight edge overall.

4. Piastri (PIA)
- 1st Stint (Medium Tires):
    Piastris median time on mediums is just above 1:38.0, indicating a solid but slower pace compared to Norris and Verstappen.
    The IQR for Piastris first stint is moderate, showing some variability in his lap times.
- 2nd Stint (Hard Tires):
    On the hards, Piastris median time improves slightly but remains above 1:37.5, which is slower than Norris and Verstappen.
    His IQR is also relatively wide, showing less consistency on the hards compared to the other drivers.

Summary and Insights:
- Norris was the standout driver in terms of both pace and consistency across both stints. His narrow IQR and fast median times on both compounds suggest that he had an optimal race setup and managed his tires effectively.
- Leclerc improved significantly on the hard tires, with a much faster median pace in the second stint. However, his wide IQR in both stints points to inconsistent lap times, possibly due to external factors like traffic or tire degradation.
- Verstappen showed a decent pace on mediums and a stronger, consistent performance on hards, where he closely matched Norriss pace. His adaptability to the hard tires allowed him to compete well, but he couldnt match Norriss early pace on the mediums.
- Piastri was the slowest of the four, with median lap times consistently higher than the others in both stints. His pace and consistency were solid but not at the same level as Norris and Verstappen, particularly on the hard tires.

Overall Takeaway:
- Norris had the best overall race pace and consistency across tire compounds, giving him a strategic advantage.
- Leclerc had a competitive second stint but was hampered by inconsistent lap times in both stints.
- Verstappen showed reliable performance, especially on hard tires, making him competitive in the later stages.
- Piastri was steady but didnt have the pace to compete with the leaders
