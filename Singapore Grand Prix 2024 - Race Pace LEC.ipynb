# Compare the race pace of Leclerc with Norris, Verstappen and Piastri

```python
import fastf1
from fastf1 import plotting
from fastf1 import utils

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
race = fastf1.get_session(2024, 'Singapore', 'R')
race.load()
```

```python
race.event
```

```python
lec = race.laps.pick_driver('LEC')
nor = race.laps.pick_driver('NOR')
ver = race.laps.pick_driver('VER')
pia = race.laps.pick_driver('PIA')
```

```python
fastf1.plotting.setup_mpl(mpl_timedelta_support=True)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(ver['LapNumber'], lec['LapTime'], color='red')
ax.plot(nor['LapNumber'], nor['LapTime'], color='orange')
ax.plot(ver['LapNumber'], ver['LapTime'], color='blue')
ax.plot(ver['LapNumber'], pia['LapTime'], color='yellow')
ax.set_title("LEC vs NOR vs VER vs PIA")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
plt.show()
```

Red Line (LEC):
LEC has a high lap time in the early laps but quickly stabilizes around 1:39.
After the pit stop (arount lap 32), LEC’s lap times return to a similar range as before, though with some variability towards the end.

Blue Line (VER):
VER starts with consistently low lap times and gradually lowers his lap times to around 1:36–1:37.
Following the spike (the pit stop), VER maintains a consistent pace, mostly with lap times between 1:36 and 1:37, indicating strong, steady performance.

Orange Line (NOR):
NOR has a similar trend as VER, with a steady start around 1:39 before lowering his lap times slightly.
NOR’s lap times after the spike remain stable but slightly higher compared to VER, showing he is consistently slower by a small margin.

Yellow Line (PIA):
PIA shows similar early lap times as NOR and VER, with lap times hovering around 1:39 initially.
Post-spike, PIA’s lap times stabilize close to those of NOR, though with some variability near the end of the race.

Key Observations:
Pit Stops/Incidents: All drivers have visible spikes around laps 30–35. These large increases in lap times likely indicate pit stops or other incidents affecting their lap times.
Consistency: VER (blue line) appears to be the most consistent after the initial laps and has the lowest average lap time, suggesting he was the fastest of the four drivers over the race.

Race Phases:
Start: All drivers have a slightly slower first lap as they settle into the race.
Middle: Between laps 5 and 30, VER (blue) and NOR (orange) show slight improvement, maintaining a consistent pace around 1:36–1:39, while LEC and PIA have occasional fluctuations.
End: After the spikes around lap 30-35, all drivers settle into a consistent pace, but VER appears to maintain the lead with consistently lower lap times than the others.

Summary:
VER (blue) likely had the strongest race pace and showed the most consistency across the laps.
NOR (orange) and PIA (yellow) followed a similar pattern but were generally slower than VER.
LEC (red) had more variability in lap times, suggesting potential challenges with consistency or additional pit stops or incidents.
This graph shows how each driver’s pace evolved throughout the race, with VER appearing to dominate based on consistent, faster lap times.
